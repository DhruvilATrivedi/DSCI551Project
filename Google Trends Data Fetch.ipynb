```python
import pytrends
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pytrends.request import TrendReq

pytrend = TrendReq()
print(pytrend)
```

```python
google_top_keywords_list = ['Grammys 2021', 'Drew Brees', 'Megan Thee Stallion', 'Dua Lipa', 'BTS', 'Justin Thomas', 
               'Doja Cat', "Zack Snyder's Justice League", 'Ides of March', 'Oscar Nominations 2021', 'Bruno Mars', 'Marvin Hagler', 
               'Kirk Franklin', 'Billie Eilish', 'Warriors', 'Juventus', 'Helen Storey', 'Bad Bunny', 'Blaze Pizza', 'CBS Sports', 
               'Miranda Lambert', 'Brittany Howard', 'Maren Morris', 'Time', 'Jennifer Lopez']
len(google_top_keywords_list)
```

```python
twitter_top_keywords_list = ['#MondayMotivation', 'Beyoncé', 'Catholic Church', '#GRAMMYs', 'Megan', 'Woody Allen', 'Lil Baby', 
                            'Patriots', 'Vatican', 'Doja', 'Mank', 'Chivas', 'Corey Davis', '#OscarNoms', 
                            'namjoon', 'Sound of Metal', '#mondaythoughts', 'Julius Caesar', 'Silk Sonic', 'AOTY', 'Best Picture', 
                            'Gen X', 'John Oliver', 'Pats', 'Lakeith']
len(twitter_top_keywords_list)
```

```python
keywords_list = google_top_keywords_list + twitter_top_keywords_list
len(keywords_list)
```

```python
x = []
for i in range(len(keywords_list)):
    x.append(("{0}".format(i+1), keywords_list[i]))
x
```

```python
df = pd.DataFrame(columns = ['DateTime'])
df
```

```python
temp = pytrend.get_historical_interest(keywords = ['Grammys 2021'], year_start=2021, month_start=3, day_start=12, hour_start=0, year_end=2021, month_end=3, day_end=18, hour_end=0, cat=0, geo='', gprop='', sleep=5)
df['DateTime'] = temp.index
df
```

```python
for i in keywords_list:
    df[i] = None
df
```

```python
df_dict = {}
```

```python
df_dict.keys()
```

```python
for i in keywords_list:
    if i not in df_dict.keys():
        temp_df = pytrend.get_historical_interest(keywords = [i], year_start=2021, month_start=3, day_start=12, hour_start=0, year_end=2021, month_end=3, day_end=18, hour_end=0, cat=0, geo='', gprop='', sleep=18)
        df_dict[i] = temp_df[i]
        print(i)
df_dict
```

```python
for i in df_dict.keys():
    for j in range(len(df['DateTime'])):
        if str(df['DateTime'][j]) in df_dict[i].index:
            df[i][j] = df_dict[i][str(df['DateTime'][j])]
df
```
