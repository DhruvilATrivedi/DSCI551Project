```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
with open('tweets.json', encoding='utf-8') as f:
    data = f.readlines()
```

```python
data
```

```python
for i in range(len(data)):
    try:
        data[i] = json.loads(data[i])
    except:
        data.remove(data[i])
```

```python
data
```

```python
new_data = []
for i in data:
    new_data.append(i['data'])
new_data
```

```python
tweets_data = pd.DataFrame(new_data)
tweets_data
```

```python
tweets_data.isnull().sum()
```

```python
tweets_data = tweets_data.drop(columns = 'geo')
tweets_data
```

```python
tweets_data= tweets_data.drop(columns = 'entities')
tweets_data
```

```python
for i in range(len(tweets_data)):
    try:
        tweets_data['public_metrics'][i] = json.loads(tweets_data['public_metrics'][i])
    except:
        pass
```

```python
public_metrics = pd.DataFrame(list(tweets_data['public_metrics']))
public_metrics
```

```python
tweets_data = tweets_data.drop(columns = 'public_metrics')
tweets_data
```

```python
tweets_data.describe()
```

```python
col_list = list(tweets_data.columns)
col_list.remove('text')
col_list.remove('id')
col_list.remove('author_id')
col_list.remove('in_reply_to_user_id')

col_list
```

```python
info = {}
for i in col_list:
    info[i] = tweets_data[i].value_counts()
info
```

```python
info['lang'].index, info['lang'].values
```

```python
df = pd.DataFrame({'Language':info['lang'].index, 'Values':info['lang'].values})
df
```

```python
df.plot(kind='bar', x='Language', y='Values', figsize = (20, 15))
```

```python
info['source'].index, info['source'].values
```

```python
df = pd.DataFrame({'Source':info['source'].index, 'Values':info['source'].values})
df
```

```python
df.plot(kind='bar', x='Source', y='Values', figsize = (20, 15))
```
