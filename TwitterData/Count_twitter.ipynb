```python
import os 
import pandas as pd
import datetime
from os import listdir
from os.path import isfile, join
mypath = "Tweets_Raw_data"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

new_list = pd.date_range("2021-03-12", periods=24*6 + 1, freq="h")
tz_aware = new_list.tz_localize(tz='UTC')
tz_aware = list(tz_aware)
```

```python
onlyfiles
```

```python
FINAL_COUNT = dict.fromkeys(onlyfiles,"")
for file in onlyfiles:
    print(file)
    if file == ".DS_Store":
        continue
    df = pd.read_csv(mypath + "/" + file,lineterminator='\n').iloc[::-1]
    date_time = pd.to_datetime(df.pop('DateTime'))
    Base_date = tz_aware[0]
    i = 0
    COUNT_DICT = dict.fromkeys(tz_aware,0)
    for time in date_time:
        if time <= Base_date:

            COUNT_DICT[Base_date] += 1
        else:
            i+=1
            Base_date = tz_aware[i]
    FINAL_COUNT[file] = COUNT_DICT
```
